```python
data.describe()
```

```python
data.info()
```

```python
plt.bar(data['Vehicle_Type'],data['Number_of_Casualties'] )
plt.xlabel('Vehicle_Type')
plt.ylabel('Number_of_Casualties')
plt.title('No of casualties wrt vehicles')
plt.show()
```

```python
plt.bar(data['Urban_or_Rural_Area'],data['Number_of_Casualties'] )
plt.xlabel('Urban_or_Rural_Area')
plt.ylabel('Number_of_Casualties')
plt.title('No of casualties in Urban and Rural')
plt.show()
```

```python
plt.bar(data['Road_Surface_Conditions'],data['Number_of_Casualties'] )
plt.xlabel('Road_Surface_Conditions')
plt.ylabel('Number_of_Casualties')
plt.title('No of casualties based on road surface')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.pie(data["Accident_Severity"].value_counts(),autopct="%1.1f%%",labels=np.unique(data["Accident_Severity"]))
plt.title("Accident Severity")
plt.legend(title="Accident types", loc="center left", bbox_to_anchor=(1,0.5))
plt.axis("equal")
plt.show()
```
